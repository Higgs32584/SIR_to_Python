```python
def get_ics(model):
    ics= None
    if(model == "SIR" ):
        ics={"S":9999, "I":1, "R":0}
    if(model == "SIS" ):
        ics={"S":9999, "S":1}
    return ics
def get_name(model):
    types = {"births":"demography","vaccination":"vaccination"}
    a=""
    if("SIRS" not in model):
        a=model[0:3]
    else:
        a=model[0:4]
    if(model[len(a):len(model)] not in types.keys()):
        return "Interactive "+a+" model"
    else:
        return "Interactive "+a+" model with "+types[model[len(a):len(model)]]
```
